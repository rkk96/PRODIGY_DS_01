### Representing the data "Income Group By Region Wise"

```python
import matplotlib.pyplot as plt

Region_income = pd.crosstab(Final_data.IncomeGroup, Final_data.Region)

# Set the figure size before plotting
fig, ax = plt.subplots(figsize=(15, 8))

# Plot the cross-tabulation table
Region_income.plot(kind='bar', ax=ax)

# Add labels and title
plt.xlabel('Income Group')
plt.ylabel('Region')
plt.title('Income Group By Region Wise')
plt.legend(title='Regions')
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Specify the columns you want to select
selected_columns = ['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
       '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']

# Create the DataFrame with selected columns
years = Final_data[selected_columns]
years
```

##Population Over the Years

```python
fig, ax = plt.subplots(figsize=(20, 8))
colors = plt.cm.viridis(np.linspace(0, 3, len(years)))

years.sum().plot(kind='bar',color= colors)  # Summing up the population for each year across all regions and plotting
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Over the Years')
plt.xticks(rotation=45)
plt.show()
```

```python
# Creating a new data frame with year wise population and regions

population_columns = Final_data.columns[Final_data.columns.str.isnumeric()]
population_data = Final_data[['Region'] + population_columns.tolist()]

# Calculate the total population for each region across all years
region_population = population_data.groupby('Region').sum()

# verifying the data

population_data

```

```python
region_population
```

##Population Over the Years by Region

```python
# Create a figure and axis object
fig, ax = plt.subplots(figsize=(18, 8))

# Plot the data
region_population.T.plot(kind='bar', stacked=True, ax=ax)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Over the Years by Region')

# Move the legend inside the plot
plt.legend(title='Region', loc='upper left')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

## Count of Countries by Region and Income Group

```python
import matplotlib.pyplot as plt

# Step 1: Group the data by 'Region' and 'IncomeGroup'
grouped_data = Final_data.groupby(['Region', 'IncomeGroup']).size().unstack()

# Step 2: Plot the data as a bar chart
plt.figure(figsize=(12, 8))
grouped_data.plot(kind='bar', stacked=True)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Count of Countries by Region and Income Group')
plt.legend(title='Income Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```
